```python
import matplotlib.pyplot as plt
import numpy as np
# независимая переменная
x = np.linspace(-100, 100)

# название осей
xlab='Ось X'
ylab='Ось Y'

# первый график
y= x**2
# легенда
leg='y = x**2'

# Задаем подписи к осям X и оси Y
plt.xlabel(xlab)
plt.ylabel(ylab)

# Задаем заголовок диаграммы
plt.title('Квадратичная функция ')

# строим графики
plt.plot(x,y,'b-',label=leg)

# задаем вывод легенды и ее расположение
plt.legend(loc='best')

plt.show()
```

```python
# независимая переменная
x = np.linspace(-np.pi*5, np.pi*5, 100)

# название осей
xlab='Ось X'
ylab='Ось Y'

# первый график
y= np.sin(x)
# легенда
leg='y = sin(x)'

# Задаем подписи к осям X и оси Y
plt.xlabel(xlab)
plt.ylabel(ylab)

# Задаем заголовок диаграммы
plt.title('Гармоническая функция')

# строим графики
plt.plot(x,y,'b-',label=leg)

# задаем вывод легенды и ее расположение
plt.legend(loc='best')

plt.show()
```

```python
fig = plt.figure(figsize=(12, 6))

x = np.linspace(-np.pi*5, np.pi*5, 100)
y = np.sin(x)
z = x**2

leg1='y = sin(x)'
leg2='y = x**2'

ax = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax.set_title('Гармоническая функция')
ax.plot(y, color='blue', label=leg1)

ax2.set_title('Квадратичная функция')
ax2.plot(z, color='black', label=leg2)

ax.legend(loc='best')
plt.legend(loc='best')

plt.show()
```

```python
fig = plt.figure(figsize=(12, 6))

x1 = np.arange(0, 10, 0.1)
x2 = np.linspace(-2, 2)
y = np.sin(x1)
z = x2**2

ax = fig.add_subplot(121)

ax.set_title('Гармоническая функция и квадратичная функция')
ax.plot(y, color='blue', label='Sine wave')
ax.plot(z, color='black', label='Cosine wave')
```

```python
# независимая переменная
t = np.linspace(0, np.pi*4)

# название осей
xlab='Ось X'
ylab='Ось Y'

# первый график
y= np.sin(t*2)
x= np.cos(t*3)

# Задаем подписи к осям X и оси Y
plt.xlabel(xlab)
plt.ylabel(ylab)

# Задаем заголовок диаграммы
plt.title('График функции, заданной параметрически')

# строим графики
plt.plot(x,y,'b-')

plt.show()
```

```python
import matplotlib as mpl

data_names = ['2012', '2013', '2014', '2015', '2016',

              '2017', '2018', '2019', '2020',

              '2021', '2022']

data_values = [4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 4900.79, 4792.54, 4806.32,

               5358.53, 6268.31, 7178.32]

dpi = 80

fig = plt.figure(dpi = dpi, figsize = (512 / dpi, 384 / dpi) )

mpl.rcParams.update({'font.size': 10})

plt.title('Статистический комитет КР ')

xs = range(len(data_names))

plt.bar([x for x in xs], [ d for d in
                                       data_values],

        width = 0.2, color = 'red', alpha = 0.7,

        zorder = 2)

plt.xticks(xs, data_names)

fig.autofmt_xdate(rotation = 25)

xlab='Наименование показателей '
ylab='Кыргызская Республика'

plt.xlabel(xlab)
plt.ylabel(ylab)

plt.show()
```

```python
data_names = ['2012', '2013', '2014', '2015', '2016',

              '2017', '2018', '2019', '2020',

              '2021', '2022']

data_values = [4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 4900.79, 4792.54, 4806.32,

               5358.53, 6268.31, 7178.32]



dpi = 80

fig = plt.figure(dpi = dpi, figsize = (512 / dpi, 384 / dpi) )

mpl.rcParams.update({'font.size': 9})



plt.title('Статистический комитет КР (%)')



xs = range(len(data_names))



plt.pie(

    data_values, autopct='%.1f', radius = 1.1,

    explode = [0.15] + [0 for _ in range(len(data_names) - 1)] )

plt.legend(

    bbox_to_anchor = (-0.16, 0.45, 0.25, 0.25),

    loc = 'lower left', labels = data_names )

plt.show()
```
